```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# basic graph

```python
#for plotting simple graph
x =[5,10,15,20,30]
y =[4,6,8,20,12,]

plt.plot(x,y)
plt.show()
```

```python
#for color
plt.plot(x,y,color='red')
plt.show()
```

```python
#for marking in the graph
plt.plot(x,y,color='red',marker='*')
plt.show()
```

```python
#size and marker of color
plt.plot(x,y,color='red',marker='*',markeredgecolor='black',markersize=10)
plt.show()
```

```python
plt.plot(x,y,color='red',marker='*',linestyle='dashdot',markeredgecolor='black',markersize=10)
```

```python
#shortform(color,marker,linestyle)
plt.plot(x,y,'r*--')
plt.show()
```

```python
#for heading of the graph
plt.plot(x,y,'r*--')
plt.title('Heading 1',fontdict={'fontname':'comic Sans Ms','fontsize':20})
```

```python
plt.plot(x,y,'r*--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('perfect graph',fontdict={'fontname':'comic Sans Ms','fontsize':20})
plt.show()
```

```python
#labeling the axis and legend for the graph
plt.plot(x,y,'b*--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('perfect graph',fontdict={'fontname':'comic Sans Ms','fontsize':20})
plt.legend(['single element'])
plt.show()
```

```python
plt.plot(x,y,'b*--')
plt.savefig('perfectgraph.png',dpi=200)
```

# line graph

```python
import pandas as pd

df = pd.read_csv('data 2.csv')
df.head()
```

```python
#plotting multiple graphs

plt.plot(df.year,df.india,'b.-',label='india')
plt.plot(df.year,df.china,'r.-',label='china')
plt.plot(df.year,df.america,'g.-',label='america')
plt.plot(df.year,df.japan,'m.-',label='japan')

plt.xlabel('year')
plt.ylabel('price')
plt.title('Mobile price year wise',fontdict={'fontname':'comic Sans Ms','fontsize':20})

plt.legend()
plt.savefig('mobileprice.png',dpi=300)
plt.show()
```

#histogram

```python
bunch = [0,20,40,60,80,100]
x = [10,20,30,40,50,60]

plt.hist(x,bins=bunch,color='black')
plt.show()
```

# piechart

```python
import numpy as np
y = np.array([25,30,35,40])
name = ["anuja","krishn","radha","govind"]

plt.pie(y,labels=name)
plt.show()
```

```python
#Explode
nameexplode = [0,0,0,0.2]
plt.pie(y,labels=name,explode=nameexplode)
plt.show()
```

```python
plt.pie(y,labels=name,explode=nameexplode,shadow=True)
plt.savefig('pie.png',dpi=300)
plt.legend()
plt.show()
```

```python
import numpy as np
test = np.random.normal(200,50,400)
plt.boxplot(test)
plt.show()
```

```python
#multiple boxplot
import numpy as np
test1 = np.random.normal(200,50,400)
test2= np.random.normal(100,50,300)
test3= np.random.normal(150,100,300)
test4= np.random.normal(250,75,400)
data = [test1,test2,test3,test4]
plt.boxplot(data)
plt.show()
```
